```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Platform': ['Facebook', 'YouTube', 'Instagram', 'TikTok', 'Twitter'],
    'Monthly Active Users (Billion)': [2.9, 2.5, 2.0, 1.2, 0.5]
}
df = pd.DataFrame(data)

colors = ['#4267B2', '#FF0000', '#C13584', '#69C9D0', '#1DA1F2']

plt.figure(figsize=(8, 5))
plt.bar(df['Platform'], df['Monthly Active Users (Billion)'], color=colors)
plt.title('Monthly Active Users of Social Media Platforms')
plt.xlabel('Platform')
plt.ylabel('Monthly Active Users (Billion)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
user_growth = [1.0, 1.2, 1.5, 1.8, 2.1, 2.6, 3.0, 3.5]

plt.figure(figsize=(8, 5))
plt.plot(years, user_growth, marker='o', color='#4267B2', linestyle='-', linewidth=2,
         markerfacecolor='yellow', markeredgewidth=2)
plt.title('Social Media User Growth Over Time')
plt.xlabel('Year')
plt.ylabel('Users (Billions)')
plt.grid(color='#e0e0e0', linestyle='--')
plt.show()
```

```python
content_types = ['Videos', 'Photos', 'Stories', 'Text Posts', 'Others']
user_activity = [40, 25, 20, 10, 5]

colors = ['#FF5733', '#33FF57', '#3357FF', '#FFC300', '#C70039']

plt.figure(figsize=(7, 7))
plt.pie(user_activity, labels=content_types, autopct='%1.1f%%', colors=colors, startangle=140,
        wedgeprops={'edgecolor': 'black'})
plt.title('User Activity Distribution by Content Type')
plt.show()
```

```python
import numpy as np

likes = np.random.normal(200, 50, 1000)

plt.figure(figsize=(8, 5))
n, bins, patches = plt.hist(likes, bins=30, edgecolor='black')

for i, patch in enumerate(patches):
    plt.setp(patch, 'facecolor', plt.cm.viridis(i / len(patches)))

plt.title('Post Likes Distribution')
plt.xlabel('Number of Likes')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
engagement = [100, 80, 60, 50, 30]
active_users = [2.9, 2.5, 2.0, 1.2, 0.5]
platforms = ['Facebook', 'YouTube', 'Instagram', 'TikTok', 'Twitter']
colors = ['#4267B2', '#FF0000', '#C13584', '#69C9D0', '#1DA1F2']

plt.figure(figsize=(8, 5))
scatter = plt.scatter(engagement, active_users, color=colors, s=100, edgecolor='black')

for i, platform in enumerate(platforms):
    plt.annotate(platform, (engagement[i], active_users[i]), textcoords="offset points", xytext=(0, 5), ha='center')

plt.title('Engagement vs. Active Users')
plt.xlabel('Average Likes per Post')
plt.ylabel('Active Users (Billions)')
plt.grid(color='#e0e0e0', linestyle='--')
plt.show()
```

```python
likes_per_post = [100, 80, 60, 50, 30]
posts_per_user = [30, 25, 20, 15, 10]
active_users = [2.9, 2.5, 2.0, 1.2, 0.5]
platforms = ['Facebook', 'YouTube', 'Instagram', 'TikTok', 'Twitter']

plt.figure(figsize=(10, 6))
bubble_size = [user * 100 for user in active_users]
plt.scatter(likes_per_post, posts_per_user, s=bubble_size, alpha=0.5, color='purple', edgecolor='black')

for i, platform in enumerate(platforms):
    plt.annotate(platform, (likes_per_post[i], posts_per_user[i]), fontsize=10, ha='center')

plt.title('Likes per Post vs. Posts per User (Bubble Size = Active Users)')
plt.xlabel('Average Likes per Post')
plt.ylabel('Average Posts per User')
plt.grid(color='#e0e0e0', linestyle='--')
plt.show()
```

```python
regions = ['North America', 'Asia', 'Europe', 'South America', 'Africa']
data = [np.random.normal(loc=500, scale=100, size=100),
        np.random.normal(loc=1200, scale=200, size=100),
        np.random.normal(loc=700, scale=150, size=100),
        np.random.normal(loc=300, scale=80, size=100),
        np.random.normal(loc=200, scale=50, size=100)]

plt.figure(figsize=(10, 6))
plt.boxplot(data, labels=regions, patch_artist=True,
            boxprops=dict(facecolor='#ff7f0e', color='black'))
plt.title('Daily Active Users Across Regions')
plt.xlabel('Region')
plt.ylabel('Daily Active Users (Millions)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
from math import pi

categories = ['User Growth', 'Engagement Rate', 'Ad Revenue', 'Daily Active Users', 'Content Uploads']
values = [85, 70, 90, 80, 75]
values += values[:1]

num_vars = len(categories)

angles = [n / float(num_vars) * 2 * pi for n in range(num_vars)]
angles += angles[:1]

plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)
plt.xticks(angles[:-1], categories)

ax.plot(angles, values, color='purple', linewidth=2, linestyle='solid')
ax.fill(angles, values, color='plum', alpha=0.4)

plt.title('Social Media Platform Performance Metrics')
plt.show()
```

```python
categories = ['Start', 'New Signups', 'Churn', 'Reactivations', 'Marketing Impact', 'End']
values = [500, 150, -50, 80, 20, 700]

position = [values[0]]
for i in range(1, len(values)):
    position.append(position[-1] + values[i])

x_order = list(range(len(categories)))

colors = ['#2ca02c' if val >= 0 else '#d62728' for val in values]

plt.figure(figsize=(10, 6))
for i in range(1, len(categories)):
    plt.bar(x_order[i], values[i], bottom=position[i-1], color=colors[i], edgecolor='black')

    y_position = position[i-1] + (values[i] / 2)
    plt.text(x_order[i], y_position, f'{values[i]:+}', ha='center', va='center', color='black', fontsize=10)

plt.bar(x_order[0], values[0], color='#2ca02c', edgecolor='black')

plt.text(x_order[0], values[0] / 2, f'{values[0]}', ha='center', va='center', color='black', fontsize=10)

plt.xticks(x_order, categories)
plt.title('User Growth Waterfall')
plt.ylabel('User Count (Thousands)')
plt.xlabel('Activities')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
from sankeyflow import Sankey

plt.figure(figsize=(20, 10), dpi=144)

nodes = [
    [('Total User Activity', 1000)],
    [('Videos', 400), ('Photos', 300), ('Stories', 200), ('Others', 100)],
    [('Short Videos', 250), ('Live Videos', 150)],
    [('Personal Photos', 150), ('Branded Photos', 150)],
    [('Personal Stories', 100), ('Branded Stories', 100)],
    [('Blog Posts', 50), ('Polls', 50)],
]

flows = [
    ('Total User Activity', 'Videos', 400),
    ('Total User Activity', 'Photos', 300),
    ('Total User Activity', 'Stories', 200),
    ('Total User Activity', 'Others', 100),
    ('Videos', 'Short Videos', 250),
    ('Videos', 'Live Videos', 150),
    ('Photos', 'Personal Photos', 150),
    ('Photos', 'Branded Photos', 150),
    ('Stories', 'Personal Stories', 100),
    ('Stories', 'Branded Stories', 100),
    ('Others', 'Blog Posts', 50),
    ('Others', 'Polls', 50),
]

s = Sankey(
    flows=flows,
    nodes=nodes,
)

s.draw()

plt.title('User Activity Allocation by Content Type and Subcategories')
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Country': ['United States of America', 'China', 'Germany', 'India', 'Brazil', 'United Kingdom', 'France', 'Italy', 'Canada', 'Australia'],
    'User Activity': [900, 1200, 800, 1100, 700, 850, 750, 600, 500, 400]
}

df = pd.DataFrame(data)

url = "https://naciscdn.org/naturalearth/110m/cultural/ne_110m_admin_0_countries.zip"
world = gpd.read_file(url)

world = world.merge(df, how="left", left_on="ADMIN", right_on="Country")

fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.boundary.plot(ax=ax, linewidth=0.8)
world.plot(column='User Activity', ax=ax, legend=True,
           legend_kwds={'label': "User Activity (Millions)",
                        'orientation': "horizontal"},
           cmap='coolwarm')

plt.title("Social Media User Activity by Country")
plt.show()
```

```python
import pandas as pd
import plotly.express as px

data = {
    'Region': [
        'Africa', 'Africa', 'Africa',
        'Asia', 'Asia', 'Asia',
        'Europe', 'Europe', 'Europe',
        'North America', 'North America', 'North America',
        'South America', 'South America', 'South America',
        'Australia', 'Australia', 'Australia'
    ],
    'Country': [
        'South Africa', 'Nigeria', 'Kenya',
        'China', 'India', 'Japan',
        'Germany', 'France', 'Italy',
        'United States', 'Canada', 'Mexico',
        'Brazil', 'Argentina', 'Colombia',
        'Australia', 'New Zealand', 'Fiji'
    ],
    'State': [
        'Gauteng', 'Lagos', 'Nairobi',
        'Beijing', 'Maharashtra', 'Kyoto',
        'Bavaria', 'le-de-France', 'Lazio',
        'California', 'Ontario', 'Chihuahua',
        'So Paulo', 'Buenos Aires', 'Cundinamarca',
        'New South Wales', 'Auckland', 'Viti Levu'
    ],
    'User Activity (Millions)': [
        40.0, 30.0, 20.0,
        120.0, 90.0, 50.0,
        150.0, 130.0, 90.0,
        300.0, 180.0, 60.0,
        60.0, 50.0, 35.0,
        50.0, 30.0, 15.0
    ]
}

df = pd.DataFrame(data)

fig = px.sunburst(
    df,
    path=['Region', 'Country', 'State'],
    values='User Activity (Millions)',
    color='User Activity (Millions)',
    color_continuous_scale='RdYlGn',
    title='Global Social Media User Activity Distribution',
    width=700, height=500
)

fig.update_layout(
    margin=dict(t=40, l=0, r=0, b=0),
    coloraxis_colorbar=dict(
        title="User Activity (Millions)",
        tickvals=[0, 100, 200, 300, 400]
    )
)

fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
fig, axs = plt.subplots(2, 3, figsize=(18, 10))
axs = axs.flatten()

# 1. Linear Data
x_linear = np.linspace(0, 10, 100)
y_linear = 2 * x_linear + 1 + np.random.normal(0, 2, size=x_linear.shape)
linear_coeffs = np.polyfit(x_linear, y_linear, 1)
y_pred_linear = np.polyval(linear_coeffs, x_linear)
axs[0].scatter(x_linear, y_linear, label='Data')
axs[0].plot(x_linear, y_pred_linear, color='red', label='Fit')
axs[0].set_title('Linear Regression')
axs[0].legend()

# 2. Quadratic Data
x_quad = np.linspace(-10, 10, 100)
y_quad = 0.5 * x_quad**2 - 3 * x_quad + 2 + np.random.normal(0, 5, size=x_quad.shape)
quad_coeffs = np.polyfit(x_quad, y_quad, 2)
y_pred_quad = np.polyval(quad_coeffs, x_quad)
axs[1].scatter(x_quad, y_quad, label='Data')
axs[1].plot(x_quad, y_pred_quad, color='red', label='Fit')
axs[1].set_title('Quadratic Regression')
axs[1].legend()

# 3. Cubic Data
x_cubic = np.linspace(-5, 5, 100)
y_cubic = x_cubic**3 - 2*x_cubic**2 + x_cubic + 5 + np.random.normal(0, 10, size=x_cubic.shape)
cubic_coeffs = np.polyfit(x_cubic, y_cubic, 3)
y_pred_cubic = np.polyval(cubic_coeffs, x_cubic)
axs[2].scatter(x_cubic, y_cubic, label='Data')
axs[2].plot(x_cubic, y_pred_cubic, color='red', label='Fit')
axs[2].set_title('Cubic Regression')
axs[2].legend()

# 4. Polynomial Regression (Degree 4)
x_poly4 = np.linspace(-5, 5, 100)
y_poly4 = 0.1 * x_poly4**4 - 0.5 * x_poly4**3 + x_poly4**2 - 2 * x_poly4 + 3 + np.random.normal(0, 2, size=x_poly4.shape)
poly4_coeffs = np.polyfit(x_poly4, y_poly4, 4)
y_pred_poly4 = np.polyval(poly4_coeffs, x_poly4)
axs[3].scatter(x_poly4, y_poly4, label='Data')
axs[3].plot(x_poly4, y_pred_poly4, color='red', label='Fit')
axs[3].set_title('4th-Degree Polynomial Regression')
axs[3].legend()

# 5. Logarithmic Data (x > 0)
x_log = np.linspace(1, 10, 100)
y_log = 3 + 2 * np.log(x_log) + np.random.normal(0, 0.5, size=x_log.shape)
log_coeffs = np.polyfit(np.log(x_log), y_log, 1)
y_pred_log = log_coeffs[0] * np.log(x_log) + log_coeffs[1]
axs[4].scatter(x_log, y_log, label='Data')
axs[4].plot(x_log, y_pred_log, color='red', label='Fit')
axs[4].set_title('Logarithmic Regression')
axs[4].legend()

# 6. Exponential Data (ensure y > 0)
x_exp = np.linspace(0, 5, 100)
y_exp = 2 * np.exp(0.8 * x_exp) + np.random.normal(0, 2, size=x_exp.shape)
y_exp = np.where(y_exp > 0, y_exp, 0.1)  # adjust any non-positive values due to noise
exp_coeffs = np.polyfit(x_exp, np.log(y_exp), 1)
y_pred_exp = np.exp(exp_coeffs[0] * x_exp + exp_coeffs[1])
axs[5].scatter(x_exp, y_exp, label='Data')
axs[5].plot(x_exp, y_pred_exp, color='red', label='Fit')
axs[5].set_title('Exponential Regression')
axs[5].legend()

plt.tight_layout()
plt.show()
```

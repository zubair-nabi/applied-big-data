```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

np.random.seed(0)
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = np.sin(X).ravel() + np.random.normal(0, 0.2, X.shape[0])

degrees = [1, 3, 12]
titles = ['Underfitting (Degree 1)', 'Balanced Fit (Degree 3)', 'Overfitting (Degree 12)']
plt.figure(figsize=(18, 5))

for i, degree in enumerate(degrees):
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)

    model = LinearRegression()
    model.fit(X_poly, y)

    X_grid = np.arange(0, 5, 0.01).reshape(-1, 1)
    y_pred = model.predict(poly.fit_transform(X_grid))

    plt.subplot(1, 3, i + 1)
    plt.scatter(X, y, color='red', label='Data Points', s=30)
    plt.plot(X_grid, y_pred, color='blue', label=f'Polynomial Degree {degree}', linewidth=2)
    plt.title(titles[i])
    plt.xlabel('X')
    plt.ylabel('y')
    plt.ylim(-2, 2)
    plt.axhline(0, color='gray', lw=0.5, ls='--')
    plt.axvline(0, color='gray', lw=0.5, ls='--')
    plt.legend()

plt.tight_layout()
plt.show()
```

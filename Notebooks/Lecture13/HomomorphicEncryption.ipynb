```python
import numpy as np
import pandas as pd
import phe

np.random.seed(42)

data = {
    "PatientID": [1, 2, 3],
    "Age": [45, 50, 60],
    "BMI": [23.5, 28.0, 30.5],
    "GlucoseLevel": [95, 110, 120]
}

df = pd.DataFrame(data)

def calculate_risk_score(data):
    risk_score = data["Age"] + data["BMI"] + data["GlucoseLevel"]
    return risk_score

unencrypted_risk_scores = calculate_risk_score(df)
print("Unencrypted Risk Scores: ", unencrypted_risk_scores)

public_key, private_key = phe.generate_paillier_keypair()

encrypted_data = {
    "Age": [public_key.encrypt(val) for val in df["Age"]],
    "BMI": [public_key.encrypt(val) for val in df["BMI"]],
    "GlucoseLevel": [public_key.encrypt(val) for val in df["GlucoseLevel"]]
}

def encrypted_risk_score(encrypted_data, public_key):
    encrypted_risk_scores = []
    for i in range(len(encrypted_data["Age"])):
        encrypted_risk = (
            encrypted_data["Age"][i] +
            encrypted_data["BMI"][i] +
            encrypted_data["GlucoseLevel"][i]
        )
        encrypted_risk_scores.append(encrypted_risk)
    return encrypted_risk_scores

encrypted_risk_scores = encrypted_risk_score(encrypted_data, public_key)

decrypted_risk_scores = [private_key.decrypt(score) for score in encrypted_risk_scores]

print("Decrypted Risk Scores:   ", decrypted_risk_scores)
```
